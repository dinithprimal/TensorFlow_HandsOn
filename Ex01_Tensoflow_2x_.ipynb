```python
# Rank and Degree of Tensors

rank1_tensor = tf.Variable(['Test', 'Ok', 'Time'], tf.string)
rank2_tensor = tf.Variable([['Test', 'Ok'],['Tim', 'Bill']], tf.string)
```

```python
print(tf.rank(rank1_tensor))
print(tf.rank(rank2_tensor))
```

# Changing Shape of a tensor

```python
tensor1 = tf.ones([1, 2, 3])
tensor2 = tf.reshape(tensor1, [2, 3, 1])
tensor3 = tf.reshape(tensor2, [3, -1])
```

```python
print(tensor1)
print(tensor2)
print(tensor3)
```

```python
t = tf.zeros([5,5,5,5])
print(t)
```

```python
t = tf.reshape(t, 625)
print(t)
```

```python
t = tf.reshape(t, [25, -1])
print(t)
```

```python
t = tf.reshape(t, [25, 5, -1])
print(t)
```
